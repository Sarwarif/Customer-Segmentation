<H2> RFM analysis of historical sales data </H2>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
pwd()
```

```python
data = pd.read_csv("sales_data_sample.csv", encoding='unicode_escape')

data.head()
```

```python
data.info()
```

```python
data.columns
```

```python
data.head()
```

```python
# dropping unncessary columns

temp = ['ADDRESSLINE1','ADDRESSLINE2','POSTALCODE', 'TERRITORY', 'PHONE', 'CITY' , 'STATE','CONTACTFIRSTNAME', 'CONTACTLASTNAME']

data.drop(temp, axis=1, inplace=True)
```

```python
data.head()
```

```python
# Regrouping product code

data['PRODUCTINITIAL'] = data['PRODUCTCODE'].str[:3]
data.drop('PRODUCTCODE', axis=1, inplace=True)
```

```python
data.info()
```

```python
# plotting the data to gain more insight

plt.rcParams['figure.figsize'] = [18,16]
data.plot(kind="box", subplots=True, layout=(4,4), sharex=False, sharey=False)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [18, 16]
data.plot(kind='density', subplots=True, layout=(4,4), sharex=False, sharey=False)
plt.show()
```

```python
#count the number of NaN values in each 


data.isnull().sum()
```

```python
corr= data.corr()

sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns)
plt.rcParams['figure.figsize'] = [12, 10]
plt.show()

## MONTH_ID is highly correlated iwht QTR_ID and thus should be ignored
```

```python
from pandas.plotting import scatter_matrix
```

```python
scatter_matrix(data)
plt.show()
```

```python
sns.regplot(x="YEAR_ID",y="QTR_ID",data=data)
plt.show()
```

```python
# Comparing sales with each year and quarter

df_year = data.groupby(by=['YEAR_ID', 'QTR_ID']).agg({'SALES': lambda x:x.sum() })

```
